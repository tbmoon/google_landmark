```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
%matplotlib inline
```

```python
train_data        = pd.read_csv('./good_quality_image_dataset.csv')
num_of_train_data = train_data['id'].shape[0]

filename    = train_data['id'].values
url         = train_data['url'].values
landmark_id = train_data['landmark_id'].values
```

```python
print("The number of classes for the number of output of features in the final FC layer: " + str(train_data['landmark_id'].nunique()))
```

```python
x = train_data.landmark_id.value_counts()
y = x[x<200]
mean = int(sum(y.values)/len(y.values))

n, bins, patches = plt.hist(x, 100, normed = True, facecolor = 'r', alpha = 1)

plt.xlabel('#examples per class')
plt.ylabel('Norm. frequency')
plt.title('Norm. frequency vs. #examples per class')
plt.text(5000, 0.0001, 'Mean (#ex. per class<200) = {}'.format(mean), fontsize = 15)
plt.grid(True)
plt.yscale('log')
plt.savefig('1.eps')
plt.show()
```

```python
x = train_data.landmark_id.value_counts()
x = x[x<200]

n, bins, patches = plt.hist(x, 200, normed = True, facecolor = 'r', alpha = 1)

plt.xlabel('#examples per class')
plt.ylabel('Norm. frequency')
plt.title('Norm. frequency vs. #examples per class (<200)')
plt.text(40, 0.01, 'Mean (#ex. per class<200) = {}'.format(mean), fontsize = 15)
plt.grid(True)
plt.yscale('log')
plt.savefig('2.eps')
plt.show()
```
