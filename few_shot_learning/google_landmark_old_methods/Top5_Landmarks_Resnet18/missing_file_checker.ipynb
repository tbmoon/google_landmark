```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline

wrkdir = '/mnt/sw/workspace'

train_data = pd.read_csv(wrkdir + '/Google/data/train.csv')

ndata = train_data['id'].shape[0]
filename = train_data['id'].values
url = train_data['url'].values
landmark_id = train_data['landmark_id'].values

print(train_data.head())
print(ndata)

outfile = open('./missing_filelist.txt', 'w')
for i in range(ndata):
    #print(os.listdir(wrkdir + '/Google/train/'))
    input_file = wrkdir + '/Google/train/' + str(landmark_id[i]) + '/' + filename[i] +'.jpg'
    if not os.path.exists(input_file):
        print(str(i) + ': ' + input_file + ' is not exist!')
        outfile.write(str(i) + ' ' + filename[i] + ' ' + str(landmark_id[i]) + '\n')
outfile.close()

print('finished!')  
```
